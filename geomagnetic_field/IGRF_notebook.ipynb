## Import modules

```python
import pmagpy.pmag as pmag
import pmagpy.ipmag as ipmag
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate random places on globe from a uniform distribution

```python
random_places = pmag.get_unf(100)
```

## Get random places into format for igrf function
ipmag.igrf takes a list in the format [Date, Altitude, Latitude, Longitude]

```python
igrf_input_list = []
for n in range(0,len(random_places)):
    list_element = [2016.0,0,random_places[n][1],random_places[n][0]]
    igrf_input_list.append(list_element)
```

## Obtain IGRF modeled field values

```python
IGRF_output = []
IGRF_dec = []
IGRF_inc = []
for n in range(0,len(igrf_input_list)):
    IGRF_for_point = ipmag.igrf(igrf_input_list[n])
    IGRF_output.append([IGRF_for_point[0],IGRF_for_point[1],IGRF_for_point[2]])
    IGRF_dec.append(IGRF_for_point[0])
    IGRF_inc.append(IGRF_for_point[1])
```

## Plot IGRF values

```python
fignum = 1
plt.figure(num=fignum,figsize=(6,6),dpi=160)
ipmag.plot_net(fignum)
ipmag.plot_di(IGRF_dec,IGRF_inc)
```
