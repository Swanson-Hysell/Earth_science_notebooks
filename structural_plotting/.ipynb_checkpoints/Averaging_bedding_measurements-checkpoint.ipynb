First, we take the bedding measurements from the T21 stratagraphic section and put them in a table. These measurements were taken on individual planes and need to be averaged to derive an approximate tilt orientation for the entire T21 section.

```python
strike_dip = [[221,29],[199,34],[227,29],[216,25]]
```

```python
pd.DataFrame(strike_dip,columns=)
```

```python



xxx_bed_S_D=pd.read_csv('../42_47_S_Ds.txt',sep='\t',header=0)
xxx_poles_to_bed=pd.read_csv('../42_47_SD_poles.txt',sep='\t',header=0)
xxx_bed_S_D
```

The best way to take the average of planes is to calculate the poles to each plane and take a mean of that "Fisherian" distribution. The plane perpendicular to the mean is the best average of the measured bedding planes.

```python
xxx_poles_to_bed
```

```python
xxx_bedpole_dir=[]
for n in range(len(xxx_poles_to_bed)):
    Dec,Inc=xxx_poles_to_bed['pdec'][n],xxx_poles_to_bed['pinc'][n]
    xxx_bedpole_dir.append([Dec,Inc,1.])
xxx_bedpole_mean=pmag.fisher_mean(xxx_bedpole_dir)
xxx_bedpole_mean['dec']
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='stereonet')

mean_S = xxx_bedpole_mean['dec'] + 90.0
mean_D = 90.0 - xxx_bedpole_mean['inc']

ax.plane(xxx_bed_S_D['bstrike'],xxx_bed_S_D['bdip'], 'g-', linewidth=1.5)
ax.pole(xxx_bed_S_D['bstrike'],xxx_bed_S_D['bdip'], 'bo', markersize=8)
ax.plane(mean_S,mean_D, 'r-', linewidth=3)
ax.pole(mean_S,mean_D, 'r^', markersize=15)
ax.grid()

plt.show()
print mean_S,mean_D
```

The mean of the four bedding plane poles has a declination of 301.9 and an inclination of 60.7. This corresponds to a mean bedding plane of:

**Strike/Dip = 31.9/29.3 SE** 

This will be the bedding attitude used to correct each cooling unit in the T21 section.
