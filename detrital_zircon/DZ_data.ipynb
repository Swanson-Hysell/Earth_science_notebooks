```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data_url = 'http://www.geosociety.org/datarepository/2019/2019154_Table%20DR1.xlsx'
DZ_data = pd.read_excel(data_url,header=4)
```

```python
DZ_data.columns.unique
```

```python
DZ_data_AFR = np.array(DZ_data.AFR.to_list())
DZ_data_AFR = DZ_data_AFR[~np.isnan(DZ_data_AFR)]

DZ_data_SAM = np.array(DZ_data.SAM.to_list())
DZ_data_SAM = DZ_data_SAM[~np.isnan(DZ_data_SAM)]

DZ_data_PAC_NZ = np.array(DZ_data.PAC_NZ.to_list())
DZ_data_PAC_NZ = DZ_data_PAC_NZ[~np.isnan(DZ_data_PAC_NZ)]

DZ_data_NAM_CA = np.array(DZ_data.NAM_CA.to_list())
DZ_data_NAM_CA = DZ_data_NAM_CA[~np.isnan(DZ_data_NAM_CA)]

DZ_data_INDIAN = np.array(DZ_data.INDIAN.to_list())
DZ_data_INDIAN = DZ_data_INDIAN[~np.isnan(DZ_data_INDIAN)]

DZ_data_EUR = np.array(DZ_data.EUR.to_list())
DZ_data_EUR = DZ_data_EUR[~np.isnan(DZ_data_EUR)]

DZ_data_AUS = np.array(DZ_data.AUS.to_list())
DZ_data_AUS = DZ_data_AUS[~np.isnan(DZ_data_AUS)]

DZ_data_ASIA = np.array(DZ_data.ASIA.to_list())
DZ_data_ASIA = DZ_data_ASIA[~np.isnan(DZ_data_ASIA)]

DZ_data_ARAB = np.array(DZ_data.ARAB.to_list())
DZ_data_ARAB = DZ_data_ARAB[~np.isnan(DZ_data_ARAB)]

DZ_data_ANT = np.array(DZ_data.ANT.to_list())
DZ_data_ANT = DZ_data_ANT[~np.isnan(DZ_data_ANT)]

DZ_data_GRN = np.array(DZ_data.GRN.to_list())
DZ_data_GRN = DZ_data_GRN[~np.isnan(DZ_data_GRN)]
```

```python
DZ_array = np.concatenate((DZ_data_AFR,DZ_data_SAM,DZ_data_PAC_NZ,DZ_data_NAM_CA,
                           DZ_data_INDIAN,DZ_data_EUR,DZ_data_AUS,DZ_data_ASIA,
                           DZ_data_ARAB,DZ_data_ANT,DZ_data_GRN),axis=0)
```

```python
len(DZ_array)
```

```python
DZ_number, bins = np.histogram(DZ_array,bins=np.arange(0,4000,10))
```

```python
plt.hist(DZ_array,bins=bins)
plt.xlim(4000,0)
plt.show()
```
