```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

```python
initial_plume_radius = 700000 #m
initial_plume_height = 70000
K = np.pi*initial_plume_radius**2*initial_plume_height
plume_viscosity = 1E18
delta_rho_g = 200

def lateral_velocity(plume_radius):
    lateral_velocity_ms = (delta_rho_g*K**3)/(np.pi**3*plume_viscosity*plume_radius**7) #m/s
    lateral_velocity = lateral_velocity_ms*60*60*24*365 #m/yr
    return(lateral_velocity)
```

```python
K
```

```python
plume_radius_list = []
time_step_list = np.linspace(0,3000000,1001)
plume_radius = initial_plume_radius #m

for t in time_step_list:
    plume_radius_list.append(plume_radius)
    velocity = lateral_velocity(plume_radius)
    time_delta = time_step_list[1]-time_step_list[0]
    spread = velocity*time_delta
    plume_radius = plume_radius + spread
```

```python
time_Myr = time_step_list/1000000
plume_radius_km = np.array(plume_radius_list)/1000

plt.plot(time_Myr,plume_radius_km)
plt.xlabel('time (Myr)')
plt.ylabel('plume radius (kilometers)')
plt.scatter(time_Myr[18],plume_radius_km[18],c='tomato',zorder=100)
plt.text(time_Myr[18]+0.05,plume_radius_km[18]-10,'SWLLIP',c='tomato',zorder=100)
plt.scatter(time_Myr[690],plume_radius_km[690],c='tomato',zorder=100)
plt.text(time_Myr[690]+0.05,plume_radius_km[690]-20,'MCR',c='tomato',zorder=100)
plt.title('plume spread (initial thickness of 70 km and diameter of 1400 km)')
plt.savefig('plume_spread.pdf')
plt.show()
```

```python
initial_plume_radius = 500000 #m
initial_plume_height = 140000
K = np.pi*initial_plume_radius**2*initial_plume_height

def lateral_velocity(plume_radius):
    lateral_velocity_ms = (200*K**3)/(3.14**3*1E18*plume_radius**7) #m/s
    lateral_velocity = lateral_velocity_ms *60*60*24*365 #m/yr
    return(lateral_velocity)
```

```python
plume_radius_list = []
time_step_list = np.linspace(0,3000000,1001)
plume_radius = 500000 #m

for t in time_step_list:
    plume_radius_list.append(plume_radius)
    velocity = lateral_velocity(plume_radius)
    time_delta = time_step_list[1]-time_step_list[0]
    spread = velocity*time_delta
    plume_radius = plume_radius + spread
```

```python
time_Myr = time_step_list/1000000
plume_radius_km = np.array(plume_radius_list)/1000

plt.plot(time_Myr,plume_radius_km,color='black')
plt.xlabel('time (Myr)')
plt.ylabel('plume radius (kilometers)')
plt.scatter(time_Myr[18],plume_radius_km[18],c='tomato',zorder=100)
plt.text(time_Myr[18]+0.05,plume_radius_km[18]-10,'SWLLIP',c='tomato',zorder=100)
plt.scatter(time_Myr[690],plume_radius_km[690],c='tomato',zorder=100)
plt.text(time_Myr[690]+0.05,plume_radius_km[690]-20,'MCR',c='tomato',zorder=100)
plt.title('plume spread (initial thickness of 70 km and diameter of 1400 km)')
plt.savefig('plume_spread.pdf')
plt.show()
```
